```python
import numpy as np
train1_df = train1_df.replace(0, np.nan)
train1_df = train1_df.dropna(subset=['pickup_longitude', 'pickup_latitude','dropoff_longitude','dropoff_latitude'])
```

```python
train1_df = train1_df.loc[train1_df['pickup_latitude'].between(40, 42)]
train1_df = train1_df.loc[train1_df['pickup_longitude'].between(-75, -72)]
train1_df = train1_df.loc[train1_df['dropoff_latitude'].between(40, 42)]
train1_df = train1_df.loc[train1_df['dropoff_longitude'].between(-75, -72)]
```

```python
##train1_df.loc[train1_df['pickup_datetime']=='2012-04-21 04:30:42 UTC']
```

```python
train1_df['pickup']= train1_df[['pickup_longitude', 'pickup_latitude']].values.tolist()
train1_df['dropoff']= train1_df[['dropoff_longitude', 'dropoff_latitude']].values.tolist()
train1_df['points']= train1_df[['pickup', 'dropoff']].values.tolist()
```

```python
import requests
url="https://graphhopper.com/api/1/route?key=7ad529b1-0715-4a02-847b-3114d54d5793"
api_key='7ad529b1-0715-4a02-847b-3114d54d5793'


def distance(points):
  
        obj1 = { "points": points }
        resp=requests.post(url, json = obj1)
        app=resp.json()
        
        print(points)
        for dictionary in app['paths']:
            try:
            
                d=dictionary['distance']
                t=dictionary['time']
                
   # print(dictionary['distance'])
    #print(dictionary['time'])
            except KeyError:
                pass
            
        #train_df['Distance']=d
        #train_df['Time']=t
       
        
        return pd.Series([d,t])  
```

```python
train1_df[['Distance', 'Time']]=train1_df.apply(lambda row : distance(row['points']), axis = 1)
```
